```python
"""
@author: Luis M. de la Cruz [Updated on Sun Feb  1 14:30:41 2026].

Licensed under CC BY-SA 4.0 (https://creativecommons.org/licenses/by-sa/4.0/).

Trabajo realizado con el apoyo del Programa UNAM-DGAPA-PAPIME, proyectos PE101019, PE101922 y PE106025.
"""

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
import macti.vis

R = lambda t, S: np.exp(-t/S) # Funci贸n del olvido
R_d = lambda t, S: -np.exp(-t/S) / S # Derivada de la funci贸n del olvido
mf = lambda t, h, S: (R(t + h, S) - R(t, S)) / h # Aproximaci贸n de la derivada de la funci贸n del olvido

def razonDeCambio(S, h, i0, anot):
    """
    Esta funci贸n muestra las gr谩ficas de R(t), R'(t) y mf(t)
    
    Parameters
    ----------
    S: float
    Representa la intensidad del recuerdo en la funci贸n del olvido.
    
    h: float
    La separaci贸n entre los dos puntos para aproximar la derivada: (R(t+h) - R(t))/h.
    
    i0: int
    ndice para definir d贸nde se calcula la derivada: td[i0]
    
    """
    # Dominio para la evaluaci贸n de las funciones: 0,1,2,..., 7
    td = np.arange(0,8,1)
    
    # Arreglo para almacenar la aproximaci贸n de la derivada m(t) en td
    m = np.zeros(len(td)) 
    
    # Aproximaci贸n de la derivada en cada punto de td
    for i,t in enumerate(td):
        m[i] = mf(t, h, S)

    # Dominio para graficar R(t) y R'(t)
    t = np.linspace(0, 7, 100)

    # Inicializaci贸n de la figura y las subgr谩ficas
    plt.figure(figsize=(14,5))
    ax1 = plt.subplot(1,2,1)
    ax2 = plt.subplot(1,2,2)
    
    # Configuraci贸n de cada gr谩fica
    ax1.set_title('Funci贸n del olvido: $R(t) = e^{-t/S}$')
    ax1.set_xlabel('$t$ [d铆as]')
    ax1.set_ylabel('$R(t)$')
    ax1.set_xlim(-0.5,8.5)
    ax1.set_ylim(-0.2, 1.1)
    
    ax2.set_title(r"Derivada $R^\prime(t)$ y su aproximaci贸n $m(t)$")
    ax2.set_xlabel("$t$ [d铆as]")
    ax2.set_ylabel("$m(t)$")
    ax2.set_xlim(-0.5,8.5)
    ax2.set_ylim(R_d(0.0, S)-0.5, 0.75)
    
    # C谩lculo de la recta tangente exacta en el punto td[i0]
    recta = R_d(td[i0],S)*(td-td[i0]) + R(td[i0], S) 

    # C谩lculo de la recta aproximada en un punto.
    recta_m = m[i0]*(td-td[i0]) + R(td[i0], S)

    # Graficaci贸n de R(t), las rectas, y los puntos para la aproximaci贸n.
    ax1.plot(t, R(t, S), '-', c='dimgrey', alpha =0.75, label='$R(t)$', lw=3)
    ax1.plot(td, recta, c = 'C2', lw = 3, alpha = 1.0, label='Recta tangente', zorder=1)
    ax1.plot(td, recta_m, c = 'red', lw = 1, zorder=2, label='Aproximaci贸n')
    ax1.scatter(td[i0], R(td[i0], S), fc='red', ec = 'black', alpha = 0.95, s = 50, zorder = 10)    
    ax1.scatter(td[i0]+h, R(td[i0]+h, S), c = 'black', s = 50, zorder = 5)    

    # Graficaci贸n de la curva de la derivada exacta y 
    # del valor exacto de la derivada en un punto y de su aproximaci贸n.
    ax2.plot(t, R_d(t, S), lw = 3, c = 'C2', label=r"$R^\prime(t)$", zorder = 3, )
    ax2.scatter(td[i0], R_d(td[i0], S), fc='C3', ec = 'black', alpha = 0.95, s = 50, label='Valor exacto', zorder = 10)
    ax2.scatter(td[i0], m[i0], fc = 'black', ec = 'black', alpha = 0.95, s = 50, label='Aproximaci贸n', zorder = 5)

    # Textos para el gr谩fico
    texto1 = ' = {:12.10f}'.format(R(td[i0], S))
    texto2 = ' = {:12.10f}'.format(R(td[i0]+h, S))
    texto3 = ' = {:12.10f}'.format(m[i0])
    texto4 = ' = {:12.10f}'.format(R_d(td[i0], S))
    
    # Mostrar valores de las aproximaciones
    if anot:
        ax1.annotate(text = '$R(t)$' + texto1, xy = (td[i0], R(td[i0], S)), xycoords='data',
                     xytext = (0.25, 0.5), textcoords='axes fraction', color ='red', fontsize=16, 
                     arrowprops=dict(arrowstyle='->', color='red', shrinkA = 5, shrinkB = 5, 
                                     connectionstyle="angle3,angleA=0,angleB=90"))
        ax1.annotate(text = '$R(t+h)$' + texto2, xy = (td[i0]+h, R(td[i0]+h, S)), xycoords='data',
                     xytext = (0.25, 0.3), textcoords='axes fraction', color ='black', fontsize=16,
                     arrowprops=dict(arrowstyle='->', color='black', shrinkA = 5, shrinkB = 5, 
                                     connectionstyle="angle3,angleA=90,angleB=0"))         
        ax2.annotate(text = '$m($' + str(i0) + '$)$' + texto3, xy = (td[i0], m[i0]), xycoords='data',
                     xytext = (0.25, 0.5), textcoords='axes fraction', color='black', fontsize=16,
                     arrowprops=dict(arrowstyle='->', color='black', shrinkA = 5, shrinkB = 5, 
                                     connectionstyle="angle3,angleA=90,angleB=0"))    
        ax2.annotate(text = r"$R^\prime($' + str(i0) + '$)$" + texto4, xy = (td[i0], R_d(td[i0], S)), xycoords='data',
                     xytext = (0.25, 0.4), textcoords='axes fraction', color='C3', fontsize=16,
                     arrowprops=dict(arrowstyle='->', color='C3', shrinkA = 5, shrinkB = 5, 
                                     connectionstyle="angle3,angleA=0,angleB=90"))
    
    # Configuraci贸n final de las gr谩ficas.
    ax2.legend(ncol=3, frameon=True)
    ax2.grid()
    ax1.legend(ncol=3, frameon=True)
    ax1.grid()
    plt.tight_layout()
    plt.show()
    if anot:
        print("R({0:}) = {1:12.10f},  m({0:}) = {2:12.10f},  R'({0:}) = {3:12.10f}".format(i0, R(td[i0], S), m[i0], R_d(td[i0], S)))
    
w = widgets.interact(razonDeCambio,
                     S = widgets.FloatSlider(min=0.5, max=5.0, step=0.5, value=0.5, description=""),
                     h = widgets.FloatSlider(min=0.1, max=1.0, step=0.1, value=1.0, description=""),
                     i0 = widgets.IntSlider(min=0, max=6, step=1, value=1, description=""),
                     anot = widgets.Checkbox(value=False, description='Muestra valores', disabled=False,indent=False)
                    )
display(w)
```
